```python
import pandas as pd
import numpy as np
import gensim
```

```python
from gensim import downloader as api
wv = api.load("word2vec-google-news-300")
```

```python
data = pd.read_csv("email_classification.csv")
data.head(20)
```

```python
mailbox = data['email']
```

```python
mailbox[9]
```

```python
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words("english")
```

```python
pure_corpus = []
for i in range(len(mailbox)):
    refine = re.sub('[^A-Za-z]',' ',mailbox[i])
    refine = refine.lower()
    pure_corpus.append(refine)
    
```

```python
pure_corpus[0]
```

```python
main_sen = []
for i in range(len(pure_corpus)):
    words = nltk.word_tokenize(pure_corpus[i])
    main_words =[]
    for word in words:
        if(word not in stop):
            main_words.append(word)
    main_sen.append(main_words)
    
```

```python
main_sen
```

```python
mail_vec =[]
for sentence in main_sen:
    features = np.zeros(300)
    for i in range(len(sentence)):
        temp = wv[sentence[i]]
        for j in range(300):
            features[j] = features[j] + temp[j]
    mail_vec.append(features)
mail_vec = np.array(mail_vec)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(data['label'])
X = mail_vec
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)*100
```
